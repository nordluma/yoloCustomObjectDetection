```python
# Imports
import cv2 as cv
import numpy as np
```

```python
cap = cv.VideoCapture('turtle.mp4')

classes = ['Turtle']

# Laddar in modellen som tränats och cfg filen
net = cv.dnn.readNetFromDarknet('yolov3_testing.cfg', 'yolov3_training_final.weights')
net.setPreferableBackend(cv.dnn.DNN_BACKEND_OPENCV)
net.setPreferableTarget(cv.dnn.DNN_TARGET_CPU)

frame_width = int(cap.get(3))
frame_height = int(cap.get(4))
 
# Definiera codec och skappa en VideoWriter objekt. Resultaten sparas som en 'result.avi' fil
out = cv.VideoWriter('result.avi',cv.VideoWriter_fourcc('M','J','P','G'), 10, (frame_width,frame_height))

def find_objects(outputs, img):
    hT, wT, cT = img.shape
    bbox = []
    class_ids = []
    confidences = []
    
    for output in outputs:
        for detection in output:
            scores = detection[5:]
            class_id = np.argmax(scores)
            confidence = scores[class_id]
            
            if(confidence > 0.5):
                w, h = int(detection[2] * wT), int(detection[3] * hT)
                x, y = int((detection[0] * wT) - w/2), int((detection[1] * hT) - h/2)
                bbox.append([x, y, w, h])
                class_ids.append(class_id)
                confidences.append(float(confidence))
                
    indexes = cv.dnn.NMSBoxes(bbox, confidences, 0.5, 0.3)
    
    for i in indexes:
        i = i[0]
        box = bbox[i]
        x, y, w, h = box[0], box[1], box[2], box[3]
        label = str(classes[class_ids[i]])
        cv.rectangle(img, (x, y), (x + w, y + h), (255, 0, 255), 2)
        cv.putText(img, f'{label.upper()} {int(confidences[i] * 100)}%',
                  (x, y - 10), cv.FONT_HERSHEY_SIMPLEX, 0.6, (255, 0, 255), 2)

while True:
    success, img = cap.read()
    
    # Så länge som det finns frames så är while loopen true
    if(success == True):
    
        blob = cv.dnn.blobFromImage(img, 1/255, (416, 416), [0, 0, 0], 1, crop = False)
        net.setInput(blob)

        layer_names = net.getLayerNames()
        output_names = [layer_names[i[0] - 1] for i in net.getUnconnectedOutLayers()]

        outputs = net.forward(output_names)

        find_objects(outputs, img)

        out.write(img)

        cv.imshow('Image', img)
        if(cv.waitKey(1) == ord('1')):
            break
    
    else:
        break
  
    cap.release()
    out.release()
cv.destroyAllWindows()
```
